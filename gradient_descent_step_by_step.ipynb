```python
from sklearn.datasets  import make_regression
import numpy as np
```

```python
x,y = make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
#applying GD assuming slope is constant m=78.35
#and let's assume the starting value for intercept b=0
y_pred =((78.35* x)+0).reshape(4)
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label='OLS')
plt.plot(x,y_pred,color='green',label='0')
plt.legend()
plt.show()
```

```python
#claculate the slope
m=78.35
b=0

loss_slope=-2*np.sum(y- m*x.ravel()- b)
loss_slope
```

```python
# let's take learning rate =0.1,for calculating the step-size
lr=0.1

step_size=loss_slope*lr
step_size
```

```python
#calculating the new intercept
b=b-step_size
b
```

```python
y_pred1=((78.35*x)+b).reshape(4)

plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label='OLS')
plt.plot(x,y_pred1,color='green',label='b={}'.format(b))
plt.plot(x,y_pred,color='yellow',label='b=0')
plt.legend()
plt.show()
```

```python
#iteration -2
loss_slope =-2*np.sum(y- m*x.ravel()- b)
loss_slope
```

```python
step_size=loss_slope*lr
step_size
```

```python
b=b-step_size
b
```

```python
y_pred2=((78.35*x)+b).reshape(4)

plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label='OLS')
plt.plot(x,y_pred2,color='green',label='b={}'.format(b))
plt.plot(x,y_pred1,color='yellow',label='b={}'.format(b))
plt.plot(x,y_pred,color='yellow',label='b={}'.format(0))
plt.legend()
plt.show()
```

```python
# Itertion -3
loss_slope= -2 *np.sum(y-m*x.ravel()-b)
loss_slope
```

```python
step_size=loss_slope*lr
step_size
```

```python
b=b-step_size
b
```

```python
y_pred3= ((78.35*x)+b).reshape(4)

plt.figure(figsize=(10,5))
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label='OLS')
plt.plot(x,y_pred3,color='green',label='b={}'.format(b))
plt.plot(x,y_pred2,color='yellow',label='b={}'.format(b))
plt.plot(x,y_pred1,color='yellow',label='b={}'.format(b))
plt.plot(x,y_pred,color='yellow',label='b=0')
plt.legend()
plt.show()
```

#in this algo when u are far u take very big steps but coming closer the stepsize decreases.

```python
#using the loop for the same
b = -100
m = 78.35
lr = 0.1

epochs=10

for i in range(epochs):
  loss_slope=-2*np.sum(y-m*x.ravel()-b)
  b = b-(lr*loss_slope)

  y_pred = m*x+b

  plt.plot(x,y_pred)
  plt.scatter(x,y)
```
